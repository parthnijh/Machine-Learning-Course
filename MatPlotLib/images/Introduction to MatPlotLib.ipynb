```python
x=[1,2,3,4]
y=[30,40,50,60]
plt.plot(x,y)
plt.show()
```

```python
##1st method
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

```python
##2ND METHOD
fig=plt.figure()
axes=fig.add_axes([1,1,1,1])
axes.plot(x,y)
plt.show()
```

```python
## 3rd methos
fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```


## MATPLOTLIB WORKFLOW EXAMPLE

```python
##import matplotlib
%matplotlib-inline
import matplotlib.pyplot as plt


## 1 Prepare Data
x=[10,20,30,40]
y=[100,200,300,400]

##2 setup plot
fig,axes=plt.subplots(figsize=(10,10))

##3 plot data
axes.plot(x,y)

##4 customize data
axes.set(title="Simple Plot",xlabel="x axis",ylabel="y axis")

##5 save the whole figure
fig.savefig()
```
