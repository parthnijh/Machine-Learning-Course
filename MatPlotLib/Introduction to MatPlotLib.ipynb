```python
##pyplot vs oo
##if quickly then pyplot but if advanced use the OO method
```

```python
over_age_50=heartattack[heartattack["age"]>50]
over_age_50
len(over_age_50)
```

```python
over_age_50.head()
```

```python
over_age_50.plot(kind="scatter",x="age",y="chol",c="target",xlim=([45,100]))
plt.show()
```

```python
## THE OO METHOD MIXED WITH THE PYPLOT METHOD
fig,ax=plt.subplots(figsize=(10,6))
over_age_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,100])
plt.show()
```

## THE OO METHOD (OBJECT ORIENTED METHOD) FROM SCRATCH

```python

fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_age_50["age"],y=over_age_50["chol"],c=over_age_50["target"])
ax.set(title="Heart Disease And Cholestrol Levels",xlabel="Age",ylabel="Cholestrol Levels")
##adding a legend
ax.legend(*scatter.legend_elements(),title="Target")
##legend elements from the c parameter
##add a horizontal line
ax.axhline(over_age_50["chol"].mean(),linestyle="--")
plt.show()
```

```python
over_age_50
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,10),sharex=True)
## ADD DATA
scatter=ax[0].scatter(x=over_age_50["age"],y=over_age_50["chol"],c=over_age_50["target"])
##CUSTOMISE
ax[0].set(title="heart disease and cholestrol levels",ylabel="cholestrol")
## ADD A LEGEND
ax[0].legend(*scatter.legend_elements(),title="Target")
## ADD A DASH LINE
ax[0].axhline(y=over_age_50["chol"].mean(),linestyle="--")
## ADD DATA
scatter1=ax[1].scatter(x=over_age_50["age"],y=over_age_50["thalach"],c=over_age_50["target"])
## CUSTOMISE
ax[1].set(title="heart disease and max heart rate",xlabel="age",ylabel="max heart rate")
## ADD A LEGEND
ax[1].legend(*scatter.legend_elements(),title="Target")
##add mean line
ax[1].axhline(y=over_age_50["thalach"].mean(),linestyle="--")
## add the title to the figure
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold")
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_age_50["age"],y=over_age_50["chol"],c=over_age_50["target"],cmap="winter")##cmap changes color scheme
ax.set(title="Heart Disease And Cholestrol Levels",xlabel="Age",ylabel="Cholestrol Levels")
##adding a legend
ax.legend(*scatter.legend_elements(),title="Target")
##legend elements from the c parameter
##add a horizontal line
ax.axhline(over_age_50["chol"].mean(),linestyle="--")
plt.show()
```

```python
## customising the y and x axis
fig,ax=plt.subplots(2,1,figsize=(10,10),sharex=True)
## ADD DATA
scatter=ax[0].scatter(x=over_age_50["age"],y=over_age_50["chol"],c=over_age_50["target"],cmap="winter")
##CUSTOMISE
ax[0].set(title="heart disease and cholestrol levels",ylabel="cholestrol")
## ADD A LEGEND
ax[0].legend(*scatter.legend_elements(),title="Target")
## ADD A DASH LINE
ax[0].axhline(y=over_age_50["chol"].mean(),linestyle="--")
ax[0].set_xlim([50,80])
## ADD DATA
scatter1=ax[1].scatter(x=over_age_50["age"],y=over_age_50["thalach"],c=over_age_50["target"],cmap="winter")
## CUSTOMISE
ax[1].set(title="heart disease and max heart rate",xlabel="age",ylabel="max heart rate")
## ADD A LEGEND
ax[1].legend(*scatter.legend_elements(),title="Target")
##add mean line
ax[1].axhline(y=over_age_50["thalach"].mean(),linestyle="--")
ax[1].set_ylim([60,200])
ax[1].set_xlim([50,80])
## add the title to the figure
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold")
plt.show()
```
